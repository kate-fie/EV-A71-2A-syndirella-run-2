```python
import pandas as pd
import glob2
import os, re
from rdkit import Chem
```

```python
csvs = glob2.glob('csvs/*.csv')
len(csvs)
```

```python
dfs = [pd.read_csv(csv) for csv in csvs]
df = pd.concat(dfs)
df.reset_index(drop=True, inplace=True)
df.to_csv('syndirella_output_041124.csv', index=False)
df
```

```python
# Identify duplicated entries based on 'compound_set' and 'inchi_key'
duplicated_entries = df[df.duplicated(subset=['compound_set', 'inchi_key'], keep=False)]
# Show the duplicated entries
duplicated_entries
```

```python
# remove duplicated entries
df.drop_duplicates(subset=['compound_set', 'inchi_key'], keep='last', inplace=True)
len(df)
```
