```python
import pandas as pd
from rdkit import Chem
```

```python
master = pd.read_csv('../syndirella_input/master_syndirella_input_231024.csv')
master
```

```python
df = pd.read_csv('/Users/kate_fieseler/PycharmProjects/EV-A71-2A-syndirella-run-2/syndirella_output/syndirella_output_011124.csv')
df
```

```python
# get the compounds not in the master list
master['inchi_key'] = master['smiles'].apply(lambda x: Chem.MolToInchiKey(Chem.MolFromSmiles(Chem.MolToSmiles(Chem.MolFromSmiles(x), isomericSmiles=False))))
diff = master[~master['inchi_key'].isin(df['inchi_key'])]
diff
```

```python
errored = master[master['inchi_key'].isin(df[df['error_message'].str.contains('None_smiles', na=False)]['inchi_key'])] # bug that fixed
errored
```

```python
to_restart = pd.concat([diff, errored]).index
len(to_restart)
```

```python
with open('/Users/kate_fieseler/PycharmProjects/EV-A71-2A-syndirella-run-2/jobs/job.txt', 'w') as f:
    for i in to_restart:
        f.write(f'job{i}.sh\n')
```
