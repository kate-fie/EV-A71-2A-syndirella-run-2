```python
import pandas as pd
from rdkit import Chem
import re
```

```python
master = pd.read_csv('../syndirella_input/master_syndirella_input_231024.csv')
master
```

```python
master['inchi_key'] = master['smiles'].apply(lambda x: Chem.MolToInchiKey(Chem.MolFromSmiles(Chem.MolToSmiles(Chem.MolFromSmiles(x), isomericSmiles=False))))
```

```python
df = pd.read_csv('../syndirella_output/syndirella_output_311024.csv')
df
```

```python
# find difference between master and df on inchi_key
diff = master[~master['inchi_key'].isin(df['inchi_key'])]
diff
```

```python
# change A71EV2A_run5 to A71EV2A_run5_1
# write jobs
# define the file path
file_path = '/Users/kate_fieseler/PycharmProjects/EV-A71-2A-syndirella-run-2/jobs/template.sh'

# Define the string to search and the replacement string
search_string = 'NUM'
memory_string = '--mem=3GB'
replacement_memory = '--mem=10GB'

path_string = 'A71EV2A_run5'
replacement_path = 'A71EV2A_run5_1'

for i in not_finished:
    replacement_string = str(i)
    
    # Open the file, read its contents, replace the occurrences, and write it back
    with open(file_path, 'r') as file:
        file_contents = file.read()
    
    # Replace the occurrences of the search string with the replacement string
    new_contents = file_contents.replace(search_string, replacement_string)
    new_contents = new_contents.replace(memory_string, replacement_memory)
    new_contents = new_contents.replace(path_string, replacement_path)
    
    new_path = f'/Users/kate_fieseler/PycharmProjects/EV-A71-2A-syndirella-run-2/jobs/job{i}.sh'
    
    # If you want to write to a new file, you can do:
    with open(new_path, 'w') as new_file:
        new_file.write(new_contents)

print("Replacement complete.")
```
