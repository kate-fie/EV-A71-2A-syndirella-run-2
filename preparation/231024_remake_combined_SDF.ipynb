```python
import glob2
from rdkit import Chem
import pandas as pd
```

```python
mols = glob2.glob('../fragments/A71EV2A/**/*.mol', recursive=True)
len(mols)
```

```python
metadata = pd.read_csv('../fragments/A71EV2A/metadata.csv')
len(metadata)
```

```python
with Chem.SDWriter('../fragments/A71EV2A_combined_aligned.sdf') as w:
    for mol in mols:
        mol = Chem.MolFromMolFile(mol)
        w.write(mol)
w.close()
```

```python
sdf = Chem.SDMolSupplier('../fragments/A71EV2A_combined_aligned.sdf')
len(sdf)
```

```python
# make sure there are matches between 'Experiment Code' and names in SDF
sdf_names = [mol.GetProp('_Name') for mol in sdf]
hit_longcodes = metadata['Experiment code'].tolist()
metadata_dict = metadata.set_index('Code')['Experiment code'].to_dict()
# can I match all 'Experiment code' to SDF names?
for code in hit_longcodes:
    if not any(code in name for name in sdf_names):
        print('no matches', code)
        # print if multiple matches
    if len([name for name in sdf_names if code in name]) > 1:
        print('multiple matches', code)
```

```python
# look into only the specific hits in the master csv
df = pd.read_csv('../syndirella_input/master_syndirella_input_221024.csv')
hits = df['hit1'].tolist() + df['hit2'].tolist() + df['hit3'].tolist()
# drop nan
hits = [hit for hit in hits if str(hit) != 'nan']
hits = set(hits)
len(hits)
```

```python
multiples = []
for hit in hits:
    longcode = metadata_dict[hit]
    if not any(longcode in name for name in sdf_names):
        print('no matches', longcode)
    if len([name for name in sdf_names if longcode in name]) > 1:
        print('multiple matches', longcode)
        multiples.append(hit)
```

```python
# get rows of df with hit in multiples
df[df['hit1'].isin(multiples) | df['hit2'].isin(multiples) | df['hit3'].isin(multiples)]
```
