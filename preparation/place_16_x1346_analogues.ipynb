```python
%load_ext autoreload
%autoreload 2
from rdkit.Chem import PandasTools
from fragmenstein import Victor, Igor
import pandas as pd
```

```python
# definitions
def place(hits,  # list of rdkit mol objects of hits to help place
          apo_file,  # path to pdb file for protein
          work_path,
          smiles,
          name):
    Igor.init_pyrosetta()
    v = Victor(hits=hits, pdb_filename=apo_file)
    v.work_path = work_path
    v.place(smiles, long_name=name)
```

```python
df = PandasTools.LoadSDF('../designs/x1346_analogues_16.sdf')
df
```

```python
# load hit
hits = PandasTools.LoadSDF('../fragments/A71EV2A_combined.sdf')
hit = hits[hits['ID'] == 'A71EV2A-x1346_A_250_1_A71EV2A-x0526+A+147+1']
PandasTools.FrameToGridImage(hit)
```

```python
hit = [hit['ROMol'].values[0]]
apo_file = '../fragments/Ax0310a_min2.pdb'
work_path = '../designs/x1346_analogues_16_placed'
for i, row in df.iterrows():
    try: 
        place(hit, apo_file, work_path, row['smiles'], f'kate-x1346-analogues-aug-2024-{i}')
    except Exception as e:
        print(e)
        continue
```
